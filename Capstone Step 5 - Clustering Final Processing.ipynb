```python
# Cluster patient states in the test set into 250 discrete cluster centroids.
# Assign test set points to the nearest cluster centroids
```

```python
import numpy as np
import pandas as pd
from pandas import DataFrame
from sklearn.cluster import KMeans
```

```python
train_data = pd.read_csv("C:\\Users\\sara1\\Desktop\\Workingdata\\rl_train_set_unscaled.csv")
val_data = pd.read_csv("C:\\Users\\sara1\\Desktop\\Workingdata\\rl_val_set_unscaled.csv")
test_data = pd.read_csv("C:\\Users\\sara1\\Desktop\\Workingdata\\rl_test_set_unscaled.csv")
```

```python
with open("C:\\Users\\sara1\\Desktop\\Workingdata\\state_features.txt") as f:
    feat = f.read()
cluster_features = feat.split()
```

```python
KMeans
```

```python
# extract out the features from data
train_cluster_data = train_data[cluster_features]
val_cluster_data = val_data[cluster_features]
test_cluster_data = test_data[cluster_features]
```

```python
# Get the size of the dataset or object
size = train_cluster_data.memory_usage(deep=True).sum()

# Print the size in bytes
print("Size of dataset: {} bytes".format(size))
```

```python
kmeans_train = KMeans(n_clusters=100, random_state=0, verbose=True, n_init=5).fit(train_cluster_data)
```

```python
# clusters for the train data
train_clusters = kmeans_train.labels_
```

```python
# here, compute cluster centroids for val and test data
val_clusters = kmeans_train.predict(val_cluster_data)
test_clusters = kmeans_train.predict(test_cluster_data)
```

```python
train_data['state'] = train_clusters
val_data['state'] = val_clusters
test_data['state'] = test_clusters
```

```python
train_set_final = DataFrame()
train_set_final['stay_id'] = train_data['stay_id']
train_set_final['state'] = train_data['state']
train_set_final['reward'] = train_data['reward']
train_set_final['max_dose_vaso'] = train_data['vaso_input']
train_set_final['mortality'] = train_data['died_in_hosp']
```

```python
val_set_final = DataFrame()
val_set_final['stay_id'] = val_data['stay_id']
val_set_final['state'] = val_data['state']
val_set_final['reward'] = val_data['reward']
val_set_final['max_dose_vaso'] = val_data['vaso_input']
val_set_final['mortality'] = val_data['died_in_hosp']
```

```python
test_set_final = DataFrame()
test_set_final['stay_id'] = test_data['stay_id']
test_set_final['state'] = test_data['state']
test_set_final['reward'] = test_data['reward']
test_set_final['max_dose_vaso'] = test_data['vaso_input']
test_set_final['mortality'] = test_data['died_in_hosp']
```

```python
train_set_final.to_csv('C:\\Users\\sara1\\Desktop\\Workingdata\\rl_train_data_discrete.csv',index=False)
val_set_final.to_csv('C:\\Users\\sara1\\Desktop\\Workingdata\\rl_val_data_discrete.csv', index=False)
test_set_final.to_csv('C:\\Users\\sara1\\Desktop\\Workingdata\\rl_test_data_discrete.csv',index=False)
```
