```python
#Import datasets into dataframes
below65 = pd.read_csv('C:\\Users\\sara1\\Desktop\\Workingdata\\rl_below65_final_cont_noterm.csv')
above65 = pd.read_csv('C:\\Users\\sara1\\Desktop\\Workingdata\\rl_above65_final_cont_noterm.csv')
```

```python
#Replace the singular value in bicarbonate with random numbers from max and min val
column_name = 'bicarbonate'

# Generate random numbers within the specified range
min_value = 20
max_value = 36
random_numbers_below65 = np.random.randint(min_value, max_value + 1, size=len(below65))
random_numbers_above65 = np.random.randint(min_value, max_value + 1, size=len(above65))

# Replace values in the column with random numbers
below65[column_name] = random_numbers_below65
above65[column_name] = random_numbers_above65
```

```python
# now split into train/validation/test sets for below 65 dataset
unique_ids_below65 = below65['stay_id'].unique()
random.shuffle(unique_ids_below65)
train_sample_below65 = 0.7
val_sample_below65 = 0.1
test_sample_below65 = 0.2
train_num_below65 = int(len(unique_ids_below65) * 0.7)
val_num_below65 = int(len(unique_ids_below65)*0.1) + train_num_below65
train_ids_below65 = unique_ids_below65[:train_num_below65]
val_ids_below65 = unique_ids_below65[train_num_below65:val_num_below65]
test_ids_below65 = unique_ids_below65[val_num_below65:]
```

```python
#Put them into dataframes for below 65 dataset
train_set_below65 = DataFrame()
train_set_below65 = below65.loc[below65['stay_id'].isin(train_ids_below65)]

val_set_below65 = DataFrame()
val_set_below65 = below65.loc[below65['stay_id'].isin(val_ids_below65)]

test_set_below65 = DataFrame()
test_set_below65 = below65.loc[below65['stay_id'].isin(test_ids_below65)]
```

```python
av_below65 = train_set_below65[column_name].mean()
std_below65 = train_set_below65[column_name].std()
train_set_below65[column_name] = (train_set_below65[column_name] - av_below65) / std_below65
val_set_below65[column_name] = (val_set_below65[column_name] - av_below65) / std_below65
test_set_below65[column_name] = (test_set_below65[column_name] - av_below65) / std_below65
```

```python
#Scale the datasets for below 65
minimum_below65 = min(train_set_below65[column_name])
maximum_below65 = max(train_set_below65[column_name])
train_set_below65[column_name] = (train_set_below65[column_name] - minimum_below65)/(maximum_below65-minimum_below65)
val_set_below65[column_name] = (val_set_below65[column_name] - minimum_below65)/(maximum_below65-minimum_below65)
test_set_below65[column_name] = (test_set_below65[column_name] - minimum_below65)/(maximum_below65-minimum_below65)
```
